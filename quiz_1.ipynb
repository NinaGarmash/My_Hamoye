### 1. Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’. Which of these Areas had the 7th lowest sum in 2017?


```python
area_2017=df[["Y2017", "Area"]].groupby("Area").sum().sort_values(by= "Y2017", ascending=True).reset_index()
area_2017
```

Areas were sorted from lowest to highest by sum per 2017. In the 7th row(index is 6) there is Guinea-Bissau. So, **Guinea-Bissau** had the 7th lowest sum in 2017.

### 2. Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

```python
area_2017=df[["Y2017", "Area"]].groupby("Area").sum().reset_index()
area_2017
```

```python
max_area_2017 = area_2017[area_2017["Y2017"] == area_2017["Y2017"].max()]
max_area_2017
```

So, **Nigeria** had the highest sum in 2017.

### 9. What is the total number and percentage of missing data in 2014 to 3 decimal places?


```python
# Find total number of missind data in 2014
missing_2014 = df['Y2014'].isnull().sum()
missing_2014
```

```python
# Total number of rows in the DataFrame
total_rows = len(df)
# Find percentage of missing values in 2014
missing_2014_perc = (missing_2014 / total_rows) * 100
missing_2014_perc
```

So, the total number and percentage of missing data in 2014 are **1589** and **2.607** respectively.

### 11. Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?


```python
elements = df.groupby("Element").sum()
elements
```

```python
# Filter rows where 'Element' is 'Stock Variation'
stock_variation = elements[elements.index == 'Stock Variation']
stock_variation
```

We see from the table, that the highest sum of Stock Variation is in column Y2014 - value 58749.83. So, the answer is **2014**.
